```python
import numpy as np
import pandas as pd
import yfinance as yf
import warnings

warnings.filterwarnings("ignore")
pd.options.display.float_format = '{:.4%}'.format

# Date range
start = '2016-01-01'
end = '2019-12-30'

# Tickers of assets
assets = ['JCI', 'TGT', 'CMCSA', 'CPB', 'MO', 'APA', 'MMC', 'JPM',
          'ZION', 'PSA', 'BAX', 'BMY', 'LUV', 'PCAR', 'TXT', 'TMO',
          'DE', 'MSFT', 'HPQ', 'SEE', 'VZ', 'CNP', 'NI', 'T', 'BA']

assets.sort()

# Tickers of factors

factors = ['MTUM', 'QUAL', 'VLUE', 'SIZE', 'USMV']
factors.sort()

tickers = assets + factors
tickers.sort()

# Downloading data
data = yf.download(tickers, start = start, end = end)
data = data.loc[:,('Adj Close', slice(None))]
data.columns = tickers
```

```python
# Calculating returns

X = data[factors].pct_change().dropna()
Y = data[assets].pct_change().dropna()

display(X.head())
display(Y.head())
```

## 2. Estimating Mean Variance Portfolios

### 2.1 Estimating the loadings matrix.

```python
import riskfolio as rp

step = 'Forward' # Could be Forward or Backward stepwise regression
loadings = rp.loadings_matrix(X=X, Y=Y, stepwise=step)

loadings.style.format("{:.4f}").background_gradient(cmap='RdYlGn')
```

1. 행: 각 자산의 티커(symbol)를 나타냅니다. (예: APA, BA, BAX 등)
2. 열: 리스크 요인과 상수항(const)을 나타냅니다.
    * const: 회귀 분석의 상수항.
    * MTUM: 모멘텀 요인.
    * QUAL: 품질 요인.
    * SIZE: 사이즈 요인.
    * USMV: 최소 변동성 요인.
    * VLUE: 가치 요인.
3. 셀 값: 각 자산이 해당 리스크 요인에 대해 가지는 로딩(계수) 값입니다. 값이 클수록 해당 요인에 대한 민감도가 높음을 나타냅니다.
4. 색상 그라디언트: 셀 값의 크기에 따라 색상이 변합니다. 일반적으로 빨간색은 음수 값을, 초록색은 양수 값을 나타냅니다. 색상이 진할수록 절대값이 큽니다.

#### 상수항(const)의 역할

1. **회귀 모델에서의 역할**:
   - 상수항은 회귀 모델의 y절편을 나타냅니다. 이는 독립 변수의 값이 0일 때 종속 변수의 예상 값을 의미합니다.
   - 수식으로 표현하면, 회귀 모델은 다음과 같습니다:
     $$
     y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n + \epsilon
     $$
     여기서 $\beta_0$가 상수항입니다.

2. **모델의 평균 조정**:
   - 상수항은 모델의 예측 값을 종속 변수의 평균값에 맞추는 역할을 합니다. 상수항이 없다면, 모델의 예측 값이 종속 변수의 평균값과 맞지 않을 수 있습니다.

3. **편향 조정**:
   - 상수항은 모델의 편향(bias)을 조정합니다. 독립 변수와 관계없이 종속 변수에 영향을 미치는 모든 요인을 포함합니다.

```python
# Building the portfolio object
port = rp.Portfolio(returns=Y)

# Calculating optimal portfolio

# Select method and estimate input parameters:

method_mu='hist' # Method to estimate expected returns based on historical data.
method_cov='hist' # Method to estimate covariance matrix based on historical data.

port.assets_stats(method_mu=method_mu, method_cov=method_cov)

port.factors = X
port.factors_stats(method_mu=method_mu, method_cov=method_cov)

# Estimate optimal portfolio:

port.alpha = 0.05
model='FM' # Factor Model
rm = 'MV' # Risk measure used, this time will be variance
obj = 'Sharpe' # Objective function, could be MinRisk, MaxRet, Utility or Sharpe
hist = False # Use historical scenarios for risk measures that depend on scenarios
rf = 0 # Risk free rate
l = 0 # Risk aversion factor, only useful when obj is 'Utility'

w = port.optimization(model=model, rm=rm, obj=obj, rf=rf, l=l, hist=hist)

display(w.T)
```

### 2.2 Calculating the portfolio that maximizes Sharpe ratio.

#### 신뢰 수준의 정의
- 신뢰 수준은 통계적 측정에서 얼마나 신뢰할 수 있는지를 나타내는 지표입니다. 일반적으로 신뢰 수준은 1에서 α를 뺀 값으로 표현되며, α는 유의 수준이라고도 합니다.
- 예를 들어, α가 0.05이면 신뢰 수준은 95%가 됩니다. 이는 주어진 리스크 측정 방법에서 95%의 신뢰 구간 내에 결과가 있을 것이라고 예상하는 것을 의미합니다.

#### 포트폴리오 최적화에서의 역할
- 리스크 측정 방법 중에서 특히 **CVaR (Conditional Value at Risk)** 및 **EVaR (Entropic Value at Risk)**와 같은 방법들은 신뢰 수준을 필요로 합니다. 이 경우, α 값이 사용됩니다.
- **CVaR**는 VaR (Value at Risk)을 초과하는 손실의 기대값을 측정하는 방법으로, 주어진 신뢰 수준에서 손실이 발생할 확률을 고려합니다.
- **EVaR**는 리스크의 불확실성을 측정하는 방법으로, 신뢰 수준을 통해 리스크를 평가합니다.

#### 구체적인 설정
- `port.alpha = 0.05`는 신뢰 수준을 95%로 설정합니다. 이는 리스크 측정 시, 95% 신뢰 구간 내에서 결과를 얻고자 하는 것을 의미합니다.
- 예를 들어, CVaR를 사용한 경우, α가 0.05로 설정되면 95%의 신뢰 수준에서 가장 큰 손실의 평균을 측정하게 됩니다.


#### `hist = False`를 설정한 이유

`hist = False`는 최적화 과정에서 시나리오 기반의 리스크 측정 방법을 사용할 때, 역사적 시나리오를 사용하지 않도록 설정하는 것을 의미합니다. 이는 다음과 같은 이유들로 중요할 수 있습니다:

1. **예측 모델 사용**:
   - 역사적 데이터는 과거의 정보를 기반으로 하며, 이는 미래의 시장 상황을 정확히 반영하지 않을 수 있습니다. `hist = False`로 설정하면, 과거의 데이터를 사용하지 않고 예측 모델이나 현재의 시장 상황을 반영하여 리스크를 측정할 수 있습니다.

2. **시뮬레이션 기반 접근법**:
   - `hist = False`를 사용하면 몬테카를로 시뮬레이션 등 시뮬레이션 기반의 접근법을 사용하여 다양한 시나리오를 생성할 수 있습니다. 이는 다양한 가능성을 고려하여 리스크를 측정할 수 있게 합니다.

3. **리스크 측정의 유연성**:
   - 역사적 데이터에 의존하지 않음으로써, 리스크 측정 방법이 더 유연해집니다. 이는 시장 변화에 더 빠르게 대응할 수 있게 하며, 현재 시장 조건에 기반한 리스크 평가를 가능하게 합니다.

4. **극단적 사건 반영**:
   - 역사적 데이터에는 드물게 발생하는 극단적 사건이 충분히 반영되지 않을 수 있습니다. `hist = False` 설정을 통해 다양한 시나리오를 고려함으로써, 이러한 극단적 사건을 더 잘 반영할 수 있습니다.


```python
# Building the portfolio object
port = rp.Portfolio(returns=Y)

# Calculating optimal portfolio

# Select method and estimate input parameters:

method_mu='hist' # Method to estimate expected returns based on historical data.
method_cov='hist' # Method to estimate covariance matrix based on historical data.

port.assets_stats(method_mu=method_mu, method_cov=method_cov)

port.factors = X
port.factors_stats(method_mu=method_mu, method_cov=method_cov)

# Estimate optimal portfolio:

port.alpha = 0.05
model='FM' # Factor Model
rm = 'MV' # Risk measure used, this time will be variance
obj = 'Sharpe' # Objective function, could be MinRisk, MaxRet, Utility or Sharpe
hist = False # Use historical scenarios for risk measures that depend on scenarios
rf = 0 # Risk free rate
l = 0 # Risk aversion factor, only useful when obj is 'Utility'

w = port.optimization(model=model, rm=rm, obj=obj, rf=rf, l=l, hist=hist)

display(w.T)
```

### 2.3 Plotting portfolio composition

```python
# Plotting the composition of the portfolio

ax = rp.plot_pie(w=w, title='Sharpe FM Mean Variance', others=0.05, nrow=25, cmap = "tab20",
                 height=6, width=10, ax=None)
```

### 2.3 Calculate efficient frontier

```python
points = 50 # Number of points of the frontier

frontier = port.efficient_frontier(model=model, rm=rm, points=points, rf=rf, hist=hist)

display(frontier.T.head())
```

```python
# Plotting the efficient frontier

label = 'Max Risk Adjusted Return Portfolio' # Title of point
mu = port.mu_fm # Expected returns
cov = port.cov_fm # Covariance matrix
returns = port.returns_fm # Returns of the assets

ax = rp.plot_frontier(w_frontier=frontier, mu=mu, cov=cov, returns=returns, rm=rm,
                      rf=rf, alpha=0.05, cmap='viridis', w=w, label=label,
                      marker='*', s=16, c='r', height=6, width=10, ax=None)
```

```python
# Plotting efficient frontier composition

ax = rp.plot_frontier_area(w_frontier=frontier, cmap="tab20", height=6, width=10, ax=None)
```

## 3. Optimization with Constraints on Risk Factors

### 3.1 Statistics of Risk Factors

```python
# Displaying factors statistics

display(loadings.min())
display(loadings.max())
display(X.corr())
```

### 3.2 Creating Constraints on Risk Factors

```python
# Creating risk factors constraints

constraints = {'Disabled': [False, False, False, False, False],
               'Factor': ['MTUM', 'QUAL', 'SIZE', 'USMV', 'VLUE'],
               'Sign': ['<=', '<=', '<=', '>=', '<='],
               'Value': [-0.3, 0.8, 0.4, 0.8 , 0.9],
               'Relative Factor': ['', 'USMV', '', '', '']}

constraints = pd.DataFrame(constraints)

display(constraints)
```

```python
C, D = rp.factors_constraints(constraints, loadings)
```

```python
port.ainequality = C
port.binequality = D

w = port.optimization(model=model, rm=rm, obj=obj, rf=rf, l=l, hist=hist)

display(w.T)
```

```python
import statsmodels.api as sm # statsmodels는 통계 모델을 생성하고 검정하는 데 사용되는 파이썬 라이브러리

X1 = sm.add_constant(X) # 리스크 요인 데이터 X에 상수항(constant term, intercept)을 추가합니다. 이는 회귀 분석에서 y절편을 포함하기 위해 필요합니다.
y = np.matrix(returns) * np.matrix(w) # returns는 자산의 수익률 데이터프레임 w는 최적화된 포트폴리오 가중치 벡터 자산의 수익률과 가중치를 곱하여 포트폴리오의 총 수익률을 계산
results = sm.OLS(y, X1).fit() # 독립 변수 X1(리스크 요인)과 종속 변수 y(가중치가 적용된 수익률)를 사용하여 선형 회귀 모델을 생성합니다. .fit() 메서드는 모델을 데이터에 적합시켜 회귀 분석을 수행
coefs = results.params # 회귀 분석의 결과로 도출된 계수(coefficient)를 반환

print(coefs)
```

## 4. Estimating Portfolios Using Risk Factors with Other Risk Measures

### 4.1 Calculate Optimal Portfolios for Several Risk Measures.

| 약어  | 설명 | 자세한 설명 |
|-------|------|-----------|
| **MV** | Standard Deviation | 포트폴리오 수익률의 변동성을 측정합니다. 이는 수익률의 평균으로부터 각 수익률이 얼마나 떨어져 있는지를 나타내며, 변동성이 클수록 리스크가 높다고 평가됩니다. |
| **MAD** | Mean Absolute Deviation | 수익률의 평균으로부터 절대적인 편차의 평균을 측정합니다. 이는 표준 편차와 유사하지만, 평균에서 벗어난 정도를 절대값으로 계산합니다. |
| **MSV** | Semi Standard Deviation | 수익률이 평균 이하일 때의 변동성을 측정합니다. 이는 하방 리스크만을 고려하여, 평균 이상의 수익률은 무시합니다. |
| **FLPM** | First Lower Partial Moment (Omega Ratio) | 특정 기준 수익률 이하의 수익률의 평균 편차를 측정합니다. Omega Ratio는 이 값을 기준 수익률 이상의 수익률과 비교합니다. |
| **SLPM** | Second Lower Partial Moment (Sortino Ratio) | 기준 수익률 이하의 수익률 변동성을 측정합니다. Sortino Ratio는 이 변동성을 사용하여 포트폴리오의 리스크 조정 수익률을 계산합니다. |
| **CVaR** | Conditional Value at Risk | VaR(Value at Risk)을 초과하는 손실의 기대값을 측정합니다. VaR는 주어진 신뢰 수준에서 발생할 수 있는 최대 손실을 의미합니다. |
| **EVaR** | Entropic Value at Risk | 확률론적 엔트로피를 기반으로 리스크를 측정합니다. 이는 리스크의 불확실성을 측정하는 방법입니다. |
| **WR** | Worst Realization (Minimax) | 가능한 최악의 손실을 측정합니다. Minimax 접근법은 최악의 시나리오를 고려하여 의사 결정을 합니다. |
| **MDD** | Maximum Drawdown of uncompounded cumulative returns (Calmar Ratio) | 포트폴리오 가치가 최고점에서 최저점으로 떨어지는 최대 하락폭을 측정합니다. Calmar Ratio와 함께 사용됩니다. |
| **ADD** | Average Drawdown of uncompounded cumulative returns | 포트폴리오 가치의 평균 하락폭을 측정합니다. |
| **CDaR** | Conditional Drawdown at Risk of uncompounded cumulative returns | VaR와 유사하지만, 최대 낙폭(Maximum Drawdown)을 초과하는 손실의 기대값을 측정합니다. |
| **EDaR** | Entropic Drawdown at Risk of uncompounded cumulative returns | 엔트로피를 사용하여 최대 낙폭의 리스크를 측정합니다. |
| **UCI** | Ulcer Index of uncompounded cumulative returns | 포트폴리오 가치가 최고점에서 떨어진 정도와 회복 기간을 종합적으로 측정합니다. 포트폴리오의 심리적 고통을 나타내는 지표로 사용됩니다. |


```python
# Risk Measures available:
#
# 'MV': Standard Deviation.
# 'MAD': Mean Absolute Deviation.
# 'MSV': Semi Standard Deviation.
# 'FLPM': First Lower Partial Moment (Omega Ratio).
# 'SLPM': Second Lower Partial Moment (Sortino Ratio).
# 'CVaR': Conditional Value at Risk.
# 'EVaR': Entropic Value at Risk.
# 'WR': Worst Realization (Minimax)
# 'MDD': Maximum Drawdown of uncompounded cumulative returns (Calmar Ratio).
# 'ADD': Average Drawdown of uncompounded cumulative returns.
# 'CDaR': Conditional Drawdown at Risk of uncompounded cumulative returns.
# 'EDaR': Entropic Drawdown at Risk of uncompounded cumulative returns.
# 'UCI': Ulcer Index of uncompounded cumulative returns.

# port.reset_linear_constraints() # To reset linear constraints (factor constraints)

rms = ['MV', 'MAD', 'MSV', 'FLPM', 'SLPM', 'CVaR',
       'EVaR', 'WR', 'MDD', 'ADD', 'CDaR', 'UCI', 'EDaR']

w_s = pd.DataFrame([])

# When we use hist = True the risk measures all calculated
# using historical returns, while when hist = False the
# risk measures are calculated using the expected returns 
#  based on risk factor model: R = a + B * F

hist = False
for i in rms:
    w = port.optimization(model=model, rm=i, obj=obj, rf=rf, l=l, hist=hist)
    w_s = pd.concat([w_s, w], axis=1)
    
w_s.columns = rms
```
